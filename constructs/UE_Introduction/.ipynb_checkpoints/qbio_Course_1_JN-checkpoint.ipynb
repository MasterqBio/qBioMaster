**2 - Cellular membrane as a RC circuit**

2.1 - Passive membranes

Cell membranes are lipid bi-layers separating the inside and the outside of the cell. 

Theer is a negative (-70mV) voltage difference between intracellular and extracellular compartments (see below).

The membrane thus acts as an electric isolator between opposite electric charges, of capacitance C.

But proteins called ion channels are like tubes embedded and traversing the membrane, and are permeable to certain kinds of ions. Ion channels act as ohmic resistor (with resistance R or conductance g = 1/R). 

A difference in the molecular compositions (ions, proteins) of the intracellular and extracellular milieus acts as a battery.

So passive membranes are RC circuits.

2.2 - Nernst potential

Ion equilibrium and membrane potential

Suppose that you have a jar filled with water, which is separated in two
compartments by a membrane permeable only to Cl- and you add NaCl only to one of the compartments. 

What will happen ?

As time passes, Cl- ions will diffuse across the membrane (increasing the entropy of the system) and an electric potential will develop between the two compartments. The increase in potential difference creates in turn an electric field across the membrane that will oppose the increase in ionic imbalance. Eventually the system reaches a dynamic equilibrium, where diffusion of Cl- is balanced by the electric potential difference.

Nernst’s equation
\begin{equation}
V_{ion} = \frac{RT}{zF} ln( \frac{[Ion]_{out}}{[Ion]_{in}})
\end{equation}

There are several varieties of ions (K+, Na+, Cl-) 

Goldman-Hodgkin-Katz equation for the membrane potential equilibrium:

\begin{equation}
V_{membrane} = \frac{RT}{F} ln( \frac{p_{K}[K^+]_{out} + p_{Na}[Na^+]_{out} + p_{Cl}[Cl^-]_{in}}{p_{K}[K^+]_{in} + p_{Na}[Na^+]_{in} + p_{Cl}[Cl^-]_{out}})
\end{equation}

2.3 - ODE for passive membrane

The basic electrical circuit representing a leaky integrate-and-fire model consists of a capacitor C in parallel with a resistor R driven by a current I(t). 

According to Kirchoff's law of current conservation we can split the driving current into two components :
\begin{equation}
I(t) = I_{C} + I_{R}
\end{equation}

How to translate it into a differential equation ?

The resistive current : 
\begin{equation}
I_{R} = (V(t)-VI_{ion})/R 
\\
I_{R} = g(V(t)-V_{ion})
\end{equation}

The capacitive current : 
\begin{equation}
I_{C} = \frac{dq}{dt} ; 
\end{equation}

as C = q/U; 
\begin{equation}
I_{C} = C \frac{dV}{dt}
\end{equation}

So that
\begin{equation}
I(t) = C\frac{dV}{dt} +  g(V(t)-V_{ion})
\\
C\frac{dV}{dt} = -g(V(t)-V_{ion}) + I(t)
\end{equation}

What can you say about V when there isn't any current ?

What is the effect of an applied/external current ?

How does the voltage evolve when we apply a constant current "I_inj" ?


**2 - Geometric approach to ODEs**

We will first consider one-dimensional, first-order differential equations:
\begin{equation}
\frac{dx}{dt} = f(x)
\end{equation}

If x is a position along a line, dx/dt is a vector field, the velocity at each point of the line. A convenient way to represent the vector field is to plot the *phase portrait* dx/dt as function of x.

Let's consider the example 
\begin{equation}
\frac{dx}{dt}= sin(x)
\end{equation}

One can solve it analytically : 
\begin{equation}
t = ln |\frac{(csc x_{0} -cot x_{0})}{(csc x -cot x)}|
\end{equation}
\begin{equation}
csc x = \frac{1}{sin x}
\\
cot x = \frac{1}{tan x}
\end{equation}


Hard to interpret : 
- solutions ? 
- stability ?

Let's plot the phase portrait.



```python
import matplotlib.pyplot as plt ## matplotlib is a module for plotting in python
import numpy as np

x = np.arange(-1,12,.1)
dxdt = np.sin(x)
plt.plot(x,dxdt)
plt.plot([-1,12], [0., 0.], '--')
plt.plot(0.,0.,'ko')
plt.plot(2*np.pi,0.,'ko')
plt.scatter(np.pi,0.,facecolor='w',edgecolor='k')
plt.scatter(3*np.pi,0.,facecolor='w',edgecolor='k')
plt.xlabel('x', fontsize = 14)
plt.ylabel('dx/dt', fontsize = 14)
```

Convenient way to represent solutions, here fixed points that can be stable (black) or unstable (white).

Other examples to draw:

\begin{equation}
\frac{dx}{dt} = -x^3
\\
\frac{dx}{dt} = x^3
\\
\frac{dx}{dt} = x^2
\\
\frac{dx}{dt} = 0
\end{equation}


```python
import matplotlib.pyplot as plt ## matplotlib is a module for plotting in python
import numpy as np

x = np.arange(-5,5,.1)
dxdt = -x*x*x
plt.plot(x,dxdt)
plt.plot([-5,5], [0., 0.], '--')
plt.xlabel('x', fontsize = 14)
plt.ylabel('dx/dt', fontsize = 14)
```

**4- Sodium conductance : positive feedback**

Let's assume an ion channel (conductance) with increasing x_inf as a function of V and positive Vion (VNa = +50mV).

Write system of ODEs ? 
\begin{equation}
C \frac{dV}{dt} = -(g_{leak}(V(t)-V_{leak}+g_{Na}m(V)(V(t)-V_{Na})) + I(t)
\end{equation}
with
\begin{equation}
\tau_m \frac{dm}{dt} = m_{∞}-m(V)
\end{equation}

first we will take m to be instantaneous.

1) What is the analytical solution if we neglect leak (and I(t)) ?

```python
V = np.arange(-100,0.,.1)
C= 1;
gNa = 0.4;
VNa = 50;
V_AP = -50;
gL = 0.5;
VL = -70;

alpha_m = np.divide(0.182*(V+35),(1-np.exp(-(V+35)/9)));
beta_m = np.divide(-0.124*(V+35),(1-np.exp((V+35)/9)));
m_inf = alpha_m/(alpha_m+beta_m);

plt.plot(V,m_inf)

#print(m_inf)
```

```python
dVdt = -gNa*m_inf*(V-VNa)

plt.plot(V,dVdt)
plt.plot([-100,0.], [0., 0.], '--')

plt.xlabel('V', fontsize = 14)
plt.ylabel('dV/dt', fontsize = 14)
```

**6- Hodgkin-Huxley "full model"**

By recording axons from a giant squid, Hodgkin and Huxley found a voltage-dependent sodium current (described above) but also a voltage-dependent potassium current. By blocking each of the currents, they managed to retrieve the ODEs for the opening and closure of the (non-blocked) current.


\begin{equation}
C \frac{dV}{dt} = -(g_{leak}(V(t)-V_{leak}+g_{Na}m^3(V)h(V)(V(t)-V_{Na})) +g_{K}n^4(V)(V(t)-V_{K})) + I(t)
\end{equation}
with
\begin{equation}
\tau_m \frac{dm}{dt} = m_{∞}-m(V)
\\
\tau_h \frac{dh}{dt} = h_{∞}-h(V)
\\
\tau_n \frac{dn}{dt} = n_{∞}-n(V)
\end{equation}



```python
def alpha_n(a):
    return np.divide(0.02*(a-25),(1-np.exp(-(a-25)/9)))
    
def beta_n(a):
    return np.divide(-0.002*(a-25),(1-np.exp((a-25)/9)))
```

```python
delta = 0.01 # definition of Delta t
t_max = 30  # Maximal length of integration - final time
V_rest = -70;
V_1 = 50;
tstop = 10

x_num = []   # We will build here the numerical solution
t_num = []   # and here the time steps
# Initial conditions:
time = 0.
t_num.append(0.)
x_num.append(0.)

while (time < t_max):
        time += delta
        if time < tstop:
            V_m = V_1
        else :
            V_m = V_rest
        t_num.append(time)
        x_prev = x_num[-1]
        x_next = x_prev + (alpha_n(V_m)*(1-x_prev) - beta_n(V_m)*x_prev)*delta
        x_num.append(x_next)

plt.plot(t_num,x_num)

plt.xlabel('t', fontsize = 14)
plt.ylabel('n^4', fontsize = 14)
```

Hodgkin and Huxely
 applied a voltage "clamp" V_1 to the axon and recorded the current, e.g. 
\begin{equation}
I_K = g_{K}n^4(V)(V(t)-V_{K}))
\end{equation}
By dividing I_K by (V(t)-V_{K})) one can obtain g_{K}n^4(t) as a function of V_1. By integrating the activation function (e.g. n(t) ) one can show that :

\begin{equation}
n(t) = n(V_1) + (n(V_{rest}) -n(V_1)) exp (\frac{-(t-t_0)}{\tau_n(V_1)})
\end{equation}

They repeated the experiment for different values of V_1 and solved the ODEs (by hand !).
